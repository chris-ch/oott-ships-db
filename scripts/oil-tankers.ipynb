```python
import pandas
df = pandas.read_csv('data/oil-tankers.csv')
```

```python
from statsmodels.formula.api import ols
regression = ols(formula="DW ~ GT", data=df).fit()
regression.params
```

```python
regression.summary()
```

```python
outlier_test = regression.outlier_test()
outlier_test[outlier_test['bonf(p)'] < 0.01]
```
